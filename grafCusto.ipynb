```python
from sympy import *
from sympy.plotting import *
import math

x = Symbol('x')

#plot(x**2)

def cf(n, c):
  return n*c

def cespera(n, custo):
  return 2*custo/(1+n)

c = 5
ce = 10

#plot(cf(x, c), (x, 0, 5))
plot(cespera(x, ce), cf(x, c), (cespera(x, ce) + cf(x, c)), (x, 0, 5))
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(1, 7, 0.1)
def cf(n, c):
  return n*c

def cespera(n, custo):
  return custo/n
c = 20
ce = 40
plt.xlabel("# Funcion√°rios")
plt.ylabel("Custo $")
plt.plot(x, cf(x, c))
plt.plot(x, cespera(x, ce))
plt.plot(x, cf(x, c)+cespera(x, ce))
plt.show()

```
