```python
from sympy import *
from sympy.plotting import *
import math

x = Symbol('x')

#plot(x**2)

def cf(n, c):
  return n*c

def cespera(n, custo):
  return 2*custo/(1+n)

c = 5
ce = 10

#plot(cf(x, c), (x, 0, 5))
plot(cespera(x, ce), cf(x, c), (cespera(x, ce) + cf(x, c)), (x, 0, 5))
```
