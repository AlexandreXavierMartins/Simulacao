```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt

#Dados
dados = [2, 0, 2, 1, 0, 2, 1, 0, 1, 2,
         1, 2, 3, 1, 3, 1, 3, 4, 5, 1,
         2, 0, 1, 2, 1, 0, 1, 1, 0, 2,
         2, 2, 3, 2, 2, 3, 2, 3, 3, 2,
         1, 6, 0, 2, 3, 7, 0, 2, 2, 0,
         4, 1, 1, 1, 1, 8, 4, 3, 1, 4]

# Estimar a média dos dados (λ)
lambda_est = np.mean(dados)

# Ajustar a distribuição Poisson com λ = 2
x = np.arange(0, 10)  # Intervalo de valores que queremos calcular a probabilidade
poisson_pmf = poisson.pmf(x, lambda_est)

# Plotar o histograma dos dados
plt.hist(dados, bins=np.arange(-0.5, 10.5, 1), density=True, alpha=0.6, color='b', label='Dados')

# Plotar a distribuição Poisson ajustada
plt.plot(x, poisson_pmf, 'ro-', label=f'Poisson ajustado (λ={lambda_est:.2f})')

plt.xlabel('Valor')
plt.ylabel('Densidade de Probabilidade')
plt.title('Ajuste de Distribuição Poisson')
plt.legend()
plt.show()
```
